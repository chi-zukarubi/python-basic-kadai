```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
          if self.age >= 20:
            print(f"{self.name}は大人です")
          else:
            print(f"{self.name}は大人ではありません")

humans = [
    Human("太郎", 15),
    Human("花子", 25),
]

for human in humans:
    human.check_adult()
```
